```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.pyplot import cm

%matplotlib inline
```

```python
data = pd.read_csv('../datasets/test.csv', header=None)
data.columns = ['x', 'y']
data
```

```python
import os

results = os.listdir(os.getcwd())
results.remove('kmeansPlot.ipynb')
results.remove('.ipynb_checkpoints')
results.remove('kmeansResults.png')
results       
```

```python
fig, axs = plt.subplots(3, 4, figsize=(200, 170))

count = 0
for row in axs:
    for ax in row:

        df = data
        print(results[count])
        df['cluster'] = np.array(pd.read_csv(results[count], header=None)).T
        ax.set_title(results[count][:-4], fontsize=100)
        count += 1
        
        color=iter(cm.rainbow(np.linspace(0, 1, 8)))

        for i in range(8):  
            c=next(color)
            x = df[df.cluster == i]['x']
            y = df[df.cluster == i]['y']
            
            ax.scatter(x, y, color=c)
            
            
            
    plt.xlabel('X')
    plt.ylabel('Y')


plt.savefig('kmeansResults.pdf', quality=99, format='pdf')    
plt.show()
```
